# Графики для 10 контейнеров

# Графики при фиксированном infection_count = 3

```python
df = pandas.read_json('cc_10.json')
df = df[df['inf_count'] == 3]
df
```

```python
def make_pl_graphic(df, y_col, ylabel):
    fig, ax = plt.subplots()
    ax.title.set_text(ylabel)
    ax.set_xlabel('Package loss percent (%)')
    ax.set_ylabel(ylabel)
    ax.minorticks_on()
    ax.grid(which='minor', linestyle=':')
    ax.plot(df['package_loss_perc'], df[y_col])
    plt.show()
```

```python
make_pl_graphic(df, 'infected_perc', 'Infected percent (%)')
make_pl_graphic(df, 'duplicate_perc', 'Duplicate percent (%)')
make_pl_graphic(df, 'max_send_ms', 'Max send time (ms)')
make_pl_graphic(df, 'mean_send_ms', 'Mean send time (ms)')
```

## Графики при фиксированном package loss = 10 %

```python
def make_inf_graphic(df, y_col, ylabel):
    fig, ax = plt.subplots()
    ax.title.set_text(ylabel)
    ax.set_xlabel('Infection count')
    ax.set_ylabel(ylabel)
    ax.minorticks_on()
    ax.grid(which='minor', linestyle=':')
    ax.plot(df['inf_count'], df[y_col])
    plt.show()
```

```python
df = pandas.read_json('cc_10.json')
df = df[df['package_loss_perc'] == 10]
df
```

```python
make_inf_graphic(df, 'infected_perc', 'Infected percent (%)')
make_inf_graphic(df, 'duplicate_perc', 'Duplicate percent (%)')
make_inf_graphic(df, 'max_send_ms', 'Max send time (ms)')
make_inf_graphic(df, 'mean_send_ms', 'Mean send time (ms)')
```

## Трехмерная поверхность

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('cc_10.json')
```

```python
df
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator

def make_surface(df, z_title, x_title = 'package_loss_perc', y_title = 'inf_count'):
    pivoted_df = df.pivot_table(
        values=z_title,
        index=y_title,
        columns=x_title,
    )
    print(pivoted_df)
    x = pivoted_df.columns
    y = pivoted_df.index
    X,Y = np.meshgrid(x,y)
    Z = pivoted_df
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111, projection='3d')
    ax.set_xlabel(x_title)
    ax.set_ylabel(y_title)
    ax.set_zlabel(z_title)
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter('{x:.02f}')
    ax.title.set_text(z_title)
    fig.colorbar(surf, shrink=0.5, aspect=5)
    plt.show()
```

```python
make_surface(df, 'infected_perc')
```

```python
make_surface(df, 'duplicate_perc')
```

```python
make_surface(df, 'mean_send_ms')
```

```python
make_surface(df, 'max_send_ms')
```
